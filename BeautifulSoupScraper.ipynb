```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# https://github.com/metalcorebear/NRCLex
from nrclex import NRCLex

# https://github.com/aman2656/text2emotion-library
import text2emotion as te

# https://github.com/clips/pattern
from pattern.en import sentiment

url = 'https://www.theguardian.com/world/2023/may/09/french-journalist-killed-in-russian-rocket-strike-in-ukraine'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

columns = ['Title', 'URL', 'Description', 'Tags', 'Date', 'Content', 'Image']
data = []

for content in soup.find_all("div", {"class": "article-body-commercial-selector"}):
    content = content.get_text()

title = soup.find("title").get_text()
heading = soup.find("meta", property="og:title")['content']
description = soup.find("meta", property="og:description")['content']
tags = soup.find("meta", property="article:tag")['content']
section = soup.find("meta", property="article:section")['content']
image = soup.find("meta", property="og:image")['content']
published_time = soup.find("meta", property="article:published_time")['content']
site_name = soup.find("meta", property="og:site_name")['content']

text_object = NRCLex(content)
nl_emotion_scores = text_object.raw_emotion_scores
nl_affect_frequencies = text_object.affect_frequencies
te_emotion_scores = te.get_emotion(content)

print(te_emotion_scores)
print(nl_emotion_scores)
print(title)
print(heading)
print(description)
print(tags)
print(section)
print(image)
print(published_time)
print(url)
print(content)
print(site_name)

data.append([title, url, description, tags, published_time, content, image])

df = pd.DataFrame(data, columns=columns)

# print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import text2emotion as te
from nrclex import NRCLex
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from pattern.en import sentiment

URL = 'https://www.theguardian.com/international'
reqs = requests.get(URL)
soup = BeautifulSoup(reqs.text, 'html.parser')
sia = SentimentIntensityAnalyzer()

urls = []
seen = set()
for link in soup.find_all('a', {'data-link-name' : 'article'}):
    url = link.get('href')
    if url not in seen:
        seen.add(url)
        urls.append(url)

for url in urls:
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')
    for content in soup.find_all("div", {"class": "article-body-commercial-selector"}):
        content = content.get_text()
    description = soup.find("meta", property="og:description")['content']
    te_scores = te.get_emotion(description)
    text_object = NRCLex(description)
    nl_scores = text_object.raw_emotion_scores
    sia_scores = sia.polarity_scores(description)
    ps_scores = sentiment(description)
    title = soup.find("title").get_text()
    print(title)
    print(te_scores)
    print(nl_scores)
    print(sia_scores)
    print(ps_scores)
    print('---')
```
